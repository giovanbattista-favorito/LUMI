# G2 MEASUREMENT

- INPUT POWER (BEFORE MEASUREMENT): 820 mW
- STARTING POSITION: 40.2795°
- INPUT POWER (AFTER MEASUREMENT): 770 mW

```python
import numpy as np
theta = np.array([40.2795, 44.9882, 49.8729, 54.6733, 59.4774, 64.2689, 69.0135, 73.8701, 78.6726, 83.4787, 84.3080, 84.9113, 86.5032, 88.8964, 90.2690, 92.2074, 97.0083, 101.8040, 106.6023, 111.4068, 116.2072, 121.0049, 125.8007, 130.6031, 131.8867, 133.3104, 135.0764, 137.6080, 142.4006, 147.2062, 152.0097, 156.8085, 161.6089, 166.4051, 171.2185, 176.0027]) #in deg
P = np.array([561, 553, 521, 455, 377, 290, 203, 122, 57, 22, 17, 15, 13, 14, 18, 27, 71, 135, 212, 300, 382, 453, 506, 522, 527, 523, 521, 510, 462, 398, 323, 232, 156, 89, 41, 21]) #in mW
```

```python
import matplotlib.pyplot as plt

plt.grid()
plt.title("Power vs Plate angle")
plt.xlabel("Theta [deg]")
plt.ylabel("Power [mW]")
plt.plot(theta, P, 'o')
```

```python
def fit_function(x, A, B, C, D):
    return A*np.sin(B*x + C) +D
```

```python
#Fit
from scipy.optimize import curve_fit
popt, pcov = curve_fit(fit_function, theta, P, p0 = [556.50388723,   0.1,  11.77389869, 256.3122768 ])
print(popt)

plt.figure()
plt.grid()
plt.title("Power vs Plate angle (G2)")
plt.xlabel("Theta [deg]")
plt.ylabel("Power [mW]")
plt.ylim(-10, 650)
x_fit = np.linspace(40, 180, 10000)
y_fit = fit_function(x_fit, popt[0], popt[1], popt[2], popt[3])
plt.plot(theta, P, 'o')
plt.plot(x_fit, y_fit, '-')
plt.plot(87.57275728, np.min(y_fit), 'o', color = 'red', label = 'theta = 87.5728')
plt.plot(177.00370037, np.min(y_fit), 'o', color = 'lightgreen', label = 'theta = 177.0037')
plt.legend(loc = 'upper right')
```

```python
V = (np.max(y_fit)-np.min(y_fit))/(np.max(y_fit) + np.max(y_fit))
print("Visibility for G2 = {}".format(V))
```

```python
x_fit[y_fit == np.min(y_fit)]
```
